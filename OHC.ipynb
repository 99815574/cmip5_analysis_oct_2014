```python
# Calculate OHC from temperature,then culmulative sum to get total depth OHC
coords = myiris.coords.coord_names(cube, ['z', 'y', 'x'])

dims = [myiris.find_dim(cube, coord) for coord in coords]

if cube.shape != vols.shape:
    vols = iris.util.broadcast_to_shape(vols, cube.shape,dims)
    
cube = cube * vols

cube   = cube * 4184 * 1025
zcoord = myiris.coord_names(cube, 'z')[0]
zdim = myiris.find_dim(cube, zcoord)

cube.data = np.ma.cumsum(cube.data, axis=zdim)
cube.data = np.ma.getdata(cube.data)
cube.data = np.ma.masked_equal(cube.data, 0)
```

```python
def calc_stdevs(cubes):
   """
   Calculate standard deviation at each grid point.

   """
   stdev_cube = cubes[0].slices(['depth','latitude', 'longitude']).next()
   dat = myiris.return_data(cubes)

   for d in xrange(len(dat[0,:,0,0])):
       for i in xrange(len(dat[0, 0, 0, :])):
           for j in xrange(len(dat[0, 0, :, 0])):
                if ~stdev_cube.data.mask[d, j, i]:
                    
                    ydat = dat[:, d, j, i]
                    stdev_cube.data.data[d, j, i] = np.std(ydat)

   return stdev_cube
```

```python
def calc_trend(cubes):
   """
   Calculate linear trend at each grid point.

   """
   trend_cube = cubes[0].slices(['depth','latitude', 'longitude']).next()
   dat = myiris.return_data(cubes)

   for d in xrange(len(dat[0,:,0,0])):
       for i in xrange(len(dat[0, 0, 0, :])):
           for j in xrange(len(dat[0, 0, :, 0])):
                if ~trend_cube.data.mask[d, j, i]:
                    
                    ydat = dat[:, d, j, i]
                    trend.data.data[d, j, i] = np.polyfit(times,ydat,1)

   return trend_cube
```
